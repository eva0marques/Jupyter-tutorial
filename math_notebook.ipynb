#### Import des packages numpy et matplotlib.pyplot


```python
import numpy as np
import matplotlib.pyplot as plt
```

Fonction carré :
\begin{equation} 
f:\mathbb{R}\rightarrow \mathbb{R} \\
x \mapsto x^2
\end{equation}
Fonction cube :
\begin{equation} 
f: \mathbb{R}\rightarrow \mathbb{R} \\
x \mapsto x^3
\end{equation}

#### Definiton des fonctions et création de vecteurs : 

```python
def carre(x): 
    return x**2

def cube(x) :
    return x**3

x = np.arange(0,1,0.05)
y = np.arange(-2*np.pi, 2*np.pi, 0.05)
print(x)
```

#### Affichages des fonctions identité, carré et cube

```python
Carre, = plt.plot(x,carre(x), label='Carre')
Cube, = plt.plot(x,cube(x), label='Cube')
Identite, = plt.plot(x,x, label='Identite')
plt.legend(handles=[Carre, Cube, Identite])

plt.show()
```

#### Affichage des fonctions sinus et cosinus

```python
plt.figure(2)

plt.subplot(2,1,1)
plt.plot(y, np.sin(y))
plt.ylabel('sin')

plt.subplot(2,1,2)
plt.plot(y, np.cos(y))
plt.ylabel('cos')

plt.show()
```
